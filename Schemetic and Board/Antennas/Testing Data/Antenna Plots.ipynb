```python
import skrf as rf
from skrf import NetworkSet
from skrf import plotting
from skrf.plotting import save_all_figs

%matplotlib inline
from pylab import *
rf.stylely()
```

```python
antennas_dict = rf.read_all('.', contains='ant')
```

```python
antennas = NetworkSet(antennas_dict)

resonant_points = []
for a in antennas:
    resonant_points.append(argmin(abs(a.s)))
```

```python
s11_all = plt.figure()
s11_all.suptitle("S11 for all antennas")
s11_all.set_label("S11 for all antennas")
for i in range(0, len(antennas)):
    antennas[i].frequency.unit = 'GHz'
    print(antennas[i].z[resonant_points[i]])
    antennas[i].plot_s_db()

ylim([-25, 5])
    

s11_uncert = plt.figure()
s11_uncert.suptitle("S11 of all antennas with uncertainty")
ylim([-25, 5])
antennas.plot_uncertainty_bounds_s_db(label='S11');

save_all_figs('.', format=['png'])
```

```python
for a in antennas:
    a.plot_s_smith()

save_all_figs('.', format=['png'])
```
